- Openness: OPN

- Extroversion: EXT
- Agreeableness: ARG
- Neuroticism: NT
- Intelligence: INT

```python
df['e2'] = 6-df['e2'] 
df['a2'] = 6-df['a2']
df['c2'] = 6-df['c2']
df['n2'] = 6-df['n2']
df['o2'] = 6-df['o2']
df['o3'] = 6-df['o3']
df['e4'] = 6-df['e4']
df['a4'] = 6-df['a4']
df['c4'] = 6-df['c4']
df['n4'] = 6-df['n4']
df['o4'] = 6-df['o4']
```

```python
df['OPN'] = df[['o1', 'o2', 'o3', 'o4']].sum(axis = 1)
df['CSN'] = df[['c1', 'c2', 'c3', 'c4']].sum(axis = 1)
df['EXT'] = df[['e1', 'e2', 'e3', 'e4']].sum(axis = 1)
df['ARG'] = df[['a1', 'a2', 'a3', 'a4']].sum(axis = 1)
df['NT'] = df[['n1', 'n2', 'n3', 'n4']].sum(axis = 1)
df['INT'] = df[['r1', 'r2', 'r3', 'r4','r5', 'r6', 'r7', 'r8','r9', 'r10', 'r11', 'r12', 'r13']].sum(axis = 1)
```
